# 医薬品の名称について

さて、お薬の名前って、どうやって決まるんでしょうね？紛らわしいのが多くないすか？

実は、ある程度のルールがあるんです。

医薬品の名称にはステム（stem）と呼ばれる分節が含まれていて、他のものと区別できるように仕組まれています。ただし、ある薬が他の薬効を持つことが後になって知られることもあったり、ステムのルールにそぐわないものがあったりして、多少の混乱はあるようですけれども。
